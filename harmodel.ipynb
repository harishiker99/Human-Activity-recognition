```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score


import sys
import numpy
#numpy.set_printoptions(threshold=sys.maxsize)
```

```python
act = ["LAYING","SITTING","STANDING","WALKING","WALKING_DOWNSTAIRS","WALKING_UPSTAIRS",]
```

```python
#Using normalized data
train_data  = pd.read_csv("train.csv")
print(train_data)
test_data = pd.read_csv("test.csv")
print(test_data)
```

```python
#Shuffling data to remove any bias
train_data = shuffle(train_data)
test_data = shuffle(test_data)
print(train_data)
print(test_data)
```

```python
#INPUT N OUTPUT LABELS
traind = train_data.drop('Activity',axis = 1).values
trainl = train_data.Activity.values

testd = test_data.drop('Activity',axis = 1).values
testl = test_data.Activity.values
```

```python
#Encoding
from sklearn import preprocessing as pp
encoder  = pp.LabelEncoder()
encoder.fit(testl)
testle = encoder.transform(testl)
encoder.fit(trainl)
trainle = encoder.transform(trainl)
```

```python
#Machine Learning
import tensorflow as tf
from tensorflow import keras

x_train = traind
y_train = trainle

x_test = testd
y_test = testle

import sklearn.neural_network as nn

```

```python
model = nn.MLPClassifier(hidden_layer_sizes = (90,) \
                         , max_iter = 1000, alpha = 1e-4 \
                         , solver = 'sgd' , verbose = 10 \
                         , tol = 1e-19 , random_state = 1 \
                         ,learning_rate_init = .001)

nnmodel = model.fit(x_train,y_train)
```

```python
y_pred = nnmodel.predict(testd)
y_pred
```

```python
accuracy_score(testle,y_pred)
```
